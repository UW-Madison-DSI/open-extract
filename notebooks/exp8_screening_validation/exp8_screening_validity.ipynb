```python
from sklearn.metrics import cohen_kappa_score, f1_score
import pandas as pd
```

```python
df = pd.read_csv("screening_validations.csv", skiprows=1)

# Get the cols for validation
df = df[['file_name','study_within_us', "is_soybean_study", "has_yield_data", 'study_within_us.1', "is_soybean_study.1", "has_yield_data.1"]].dropna()
```

```python
# selection criteria

df["select"] = df.apply(lambda x: x["study_within_us"] and x["is_soybean_study"] and x["has_yield_data"], axis=1)
df["select.1"] = df.apply(lambda x: x["study_within_us.1"] and x["is_soybean_study.1"] and x["has_yield_data.1"], axis=1)
df
```

```python
def eval(df: pd.DataFrame, cols: list[str]) -> None:
    """Calculate the Cohen's Kappa score and f1-score for the given columns in the dataframe"""

    print("Cohen's Kappa score (Agreement):")
    for col in cols:
        kappa = cohen_kappa_score(df[col].tolist(), df[col + ".1"].tolist())
        print(f"{col}: {kappa:.2f}")
    
    print("\nF1-score (Ground-truth):")
    for col in cols:
        f1 = f1_score(df[col].tolist(), df[col + ".1"].tolist())
        print(f"{col}: {f1:.2f}")
```

Interpretation reference: https://pmc.ncbi.nlm.nih.gov/articles/PMC3900052/#:~:text=Cohen%20suggested%20the%20Kappa%20result,1.00%20as%20almost%20perfect%20agreement.

> Cohen suggested the Kappa result be interpreted as follows: values ≤ 0 as indicating no agreement and 0.01–0.20 as none to slight, 0.21–0.40 as fair, 0.41– 0.60 as moderate, 0.61–0.80 as substantial, and 0.81–1.00 as almost perfect agreement.

```python
eval(df, ["study_within_us", "is_soybean_study", "has_yield_data", "select"])
```
